```python
import pandas as pd
import mplfinance as mpf

# Load your CSV file
df = pd.read_csv('COIN-4hour-2019-01-01-to-2024-09-02.csv')

# Convert the 'timestamp' column to datetime (if not already)
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

# Drop any rows with missing or NaT timestamps
df.dropna(subset=['timestamp'], inplace=True)

# Set the 'timestamp' column as the index
df.set_index('timestamp', inplace=True)

# Verify that the index is a DatetimeIndex
if not isinstance(df.index, pd.DatetimeIndex):
    print("Index is not DatetimeIndex, converting now...")
    df.index = pd.to_datetime(df.index)

# Rename the columns to ensure they match mplfinance expectations (open, high, low, close, volume)
df.rename(columns={'open': 'Open', 'high': 'High', 'low': 'Low', 'close': 'Close', 'volume': 'Volume'}, inplace=True)

# Check if the DataFrame has the necessary columns
print(df.head())

# Plot with mplfinance
mpf.plot(df, type='candle', volume=True, title='COIN 4-hour Data', style='charles', mav=(10, 20, 50))
```
